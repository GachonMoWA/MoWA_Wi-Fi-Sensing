```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
root_Name = "Result/Result_5_31"
```

```python
ReWiS_ProtoNet_confusion = pd.read_csv(root_Name + '/few-shot/ReWiS/ProtoNet/7/500/confusion.csv', index_col=0)
ReWiS_ViT_confusion = pd.read_csv(root_Name + '/few-shot/ReWiS/ViT/7/500/confusion.csv', index_col=0)
```

```python
ReWiS_ProtoNet_confusion
```

```python
ReWiS_ViT_confusion
```

```python
categories = ['empty', 'jump', 'stand', 'walk']
ax = sns.heatmap(ReWiS_ProtoNet_confusion,annot=True,cmap='Blues')
ax.xaxis.set_ticklabels(categories)
ax.yaxis.set_ticklabels(categories)
ax.set_xlabel("Predicted Label")
ax.set_ylabel("True Label")
ax.set(title = "ProtoNet Confusion Matrix")
plt.show()
```

```python
categories = ['empty', 'jump', 'stand', 'walk']
ax = sns.heatmap(ReWiS_ViT_confusion,annot=True,cmap='Blues')
ax.xaxis.set_ticklabels(categories)
ax.yaxis.set_ticklabels(categories)
ax.set_xlabel("Predicted Label")
ax.set_ylabel("True Label")
ax.set(title = "ViT Confusion Matrix")
plt.show()
```
