```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
root_Name = "Result/Plot_data"
```

```python
UT_HAR_LeNet_train = pd.read_csv(root_Name + '/supervised/UT_HAR/LeNet/train.csv')
UT_HAR_LeNet_test = pd.read_csv(root_Name + '/supervised/UT_HAR/LeNet/test.csv')

UT_HAR_ResNet50_train = pd.read_csv(root_Name + '/supervised/UT_HAR/ResNet50/train.csv')
UT_HAR_ResNet50_test = pd.read_csv(root_Name + '/supervised/UT_HAR/ResNet50/test.csv')

UT_HAR_RNN_train = pd.read_csv(root_Name + '/supervised/UT_HAR/RNN/train.csv')
UT_HAR_RNN_test = pd.read_csv(root_Name + '/supervised/UT_HAR/RNN/test.csv')

UT_HAR_LSTM_train = pd.read_csv(root_Name + '/supervised/UT_HAR/LSTM/train.csv')
UT_HAR_LSTM_test = pd.read_csv(root_Name + '/supervised/UT_HAR/LSTM/test.csv')

UT_HAR_BiLSTM_train = pd.read_csv(root_Name + '/supervised/UT_HAR/BiLSTM/train.csv')
UT_HAR_BiLSTM_test = pd.read_csv(root_Name + '/supervised/UT_HAR/BiLSTM/test.csv')

UT_HAR_ViT_train = pd.read_csv(root_Name + '/supervised/UT_HAR/ViT/train.csv')
UT_HAR_ViT_test = pd.read_csv(root_Name + '/supervised/UT_HAR/ViT/test.csv')
```

```python
ReWiS_LeNet_train = pd.read_csv(root_Name + '/supervised/ReWiS/train_test_split_A1/LeNet/train.csv')
ReWiS_LeNet_test = pd.read_csv(root_Name + '/supervised/ReWiS/train_test_split_A1/LeNet/test.csv')

ReWiS_ResNet50_train = pd.read_csv(root_Name + '/supervised/ReWiS/train_test_split_A1/ResNet50/train.csv')
ReWiS_ResNet50_test = pd.read_csv(root_Name + '/supervised/ReWiS/train_test_split_A1/ResNet50/test.csv')

ReWiS_RNN_train = pd.read_csv(root_Name + '/supervised/ReWiS/train_test_split_A1/RNN/train.csv')
ReWiS_RNN_test = pd.read_csv(root_Name + '/supervised/ReWiS/train_test_split_A1/RNN/test.csv')

ReWiS_LSTM_train = pd.read_csv(root_Name + '/supervised/ReWiS/train_test_split_A1/LSTM/train.csv')
ReWiS_LSTM_test = pd.read_csv(root_Name + '/supervised/ReWiS/train_test_split_A1/LSTM/test.csv')

ReWiS_BiLSTM_train = pd.read_csv(root_Name + '/supervised/ReWiS/train_test_split_A1/BiLSTM/train.csv')
ReWiS_BiLSTM_test = pd.read_csv(root_Name + '/supervised/ReWiS/train_test_split_A1/BiLSTM/test.csv')

ReWiS_ViT_train = pd.read_csv(root_Name + '/supervised/ReWiS/train_test_split_A1/ViT/train.csv')
ReWiS_ViT_test = pd.read_csv(root_Name + '/supervised/ReWiS/train_test_split_A1/ViT/test.csv')
```

```python
UT_HAR_LeNet_train['model'] = 'LeNet'
UT_HAR_ResNet50_train['model'] = 'ResNet50'
UT_HAR_RNN_train['model'] = 'RNN'
UT_HAR_LSTM_train['model'] = 'LSTM'
UT_HAR_BiLSTM_train['model'] = 'BiLSTM'
UT_HAR_ViT_train['model'] = 'ViT'

UT_HAR_LeNet_test['model'] = 'LeNet'
UT_HAR_ResNet50_test['model'] = 'ResNet50'
UT_HAR_RNN_test['model'] = 'RNN'
UT_HAR_LSTM_test['model'] = 'LSTM'
UT_HAR_BiLSTM_test['model'] = 'BiLSTM'
UT_HAR_ViT_test['model'] = 'ViT'
```

```python
ReWiS_LeNet_train['model'] = 'LeNet'
ReWiS_ResNet50_train['model'] = 'ResNet50'
ReWiS_RNN_train['model'] = 'RNN'
ReWiS_LSTM_train['model'] = 'LSTM'
ReWiS_BiLSTM_train['model'] = 'BiLSTM'
ReWiS_ViT_train['model'] = 'ViT'

ReWiS_LeNet_test['model'] = 'LeNet'
ReWiS_ResNet50_test['model'] = 'ResNet50'
ReWiS_RNN_test['model'] = 'RNN'
ReWiS_LSTM_test['model'] = 'LSTM'
ReWiS_BiLSTM_test['model'] = 'BiLSTM'
ReWiS_ViT_test['model'] = 'ViT'
```

```python
df_supervised_UT_HAR_train = pd.concat([UT_HAR_LeNet_train, UT_HAR_ResNet50_train, UT_HAR_RNN_train, UT_HAR_LSTM_train, UT_HAR_BiLSTM_train, UT_HAR_ViT_train])
df_supervised_UT_HAR_test = pd.concat([UT_HAR_LeNet_test, UT_HAR_ResNet50_test, UT_HAR_RNN_test, UT_HAR_LSTM_test, UT_HAR_BiLSTM_test, UT_HAR_ViT_test])
```

```python
df_supervised_ReWiS_train = pd.concat([ReWiS_LeNet_train, ReWiS_ResNet50_train, ReWiS_RNN_train, ReWiS_LSTM_train, ReWiS_BiLSTM_train, ReWiS_ViT_train])
df_supervised_ReWiS_test = pd.concat([ReWiS_LeNet_test, ReWiS_ResNet50_test, ReWiS_RNN_test, ReWiS_LSTM_test, ReWiS_BiLSTM_test, ReWiS_ViT_test])
```

```python
sns.lineplot(data=df_supervised_UT_HAR_train, x="Epoch", y='Accuracy', hue='model', style='model')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
```

```python
sns.lineplot(data=df_supervised_UT_HAR_train, x="Epoch", y='Loss', hue='model', style='model')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
```

```python
sns.lineplot(data=df_supervised_ReWiS_train, x="Epoch", y='Accuracy', hue='model', style='model')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
```

```python
sns.lineplot(data=df_supervised_ReWiS_train, x="Epoch", y='Loss', hue='model', style='model')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
```

```python
df_best_performance_split = df_supervised_ReWiS_test.groupby('model')['Test Accuracy'].max().reset_index()
df_best_performance_split['Testing'] = "ReWiS"

df_best_performance_no = df_supervised_UT_HAR_test.groupby('model')['Test Accuracy'].max().reset_index()
df_best_performance_no['Testing'] = "UT_HAR"
```

```python
df_best_performance = pd.concat([df_best_performance_split, df_best_performance_no])
```

```python
df_best_performance
```

```python
# sns.scatterplot(data=df_supervised_ReWiS, x="model", y='Test Accuracy', hue='model', style='model')
sns.lineplot(data=df_best_performance, x="model", y='Test Accuracy', hue='Testing', marker='o')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
x_tick_labels = ['LeNet', 'ResNet50', 'RNN', 'LSTM', 'BiLSTM', 'ViT']
plt.xticks(range(len(x_tick_labels)), x_tick_labels)
plt.show()
```
